```python
import pandas as pd
```

```python
import yfinance as yf

# Define the stock symbol and the date range
symbol = 'AAPL'
start_date = '2010-01-01'
end_date = '2023-01-01'

# Download historical data
data = yf.download(symbol, start=start_date, end=end_date)

# Display the data
print(data)
```

```python
#Splitting Data into Training and Testing

data_training = pd.DataFrame(data['Close'][0:int(len(data)*0.7)])
data_testing = pd.DataFrame(data['Close'][int(len(data)*0.7):int(len(data))])

print(data_training.shape)
print(data_testing.shape)
```

```python
data_training.head()
```

```python
data_testing.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
data_training_array.shape
```

```python
data_training_array.shape[0]
```

```python
x_train = []
y_train = []

for i in range(100,data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i,0])

x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
#ML Model 
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(LSTM(units = 50, activation = 'relu', return_sequences = True,input_shape = (x_train.shape[1],1)))
model.add(Dropout(0.2))

model = Sequential()
model.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
model.add(Dropout(0.3))

model = Sequential()
model.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
model.add(Dropout(0.4))

model = Sequential()
model.add(LSTM(units = 120, activation = 'relu' ))
model.add(Dropout(0.5))

model.add(Dense(units = 1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(x_train,y_train,epochs = 50)
```

```python
model.save('my_model.h5')
```

```python
data_testing.head()
```

```python
past_100_days = data_training.tail(100)
```

```python
print(type(past_100_days))
```

```python
final_data = pd.concat([past_100_days, data_testing], ignore_index=True)
```

```python
final_data.head()
```

```python
input_data = scaler.fit_transform(final_data)
input_data
```

```python
input_data.shape
```

```python
x_test = []
y_test = []
for i in range(100,input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i,0])
```

```python
x_test, y_test = np.array(x_test), np.array(y_test)
print(x_test.shape)
print(y_test.shape)
```

```python
#Making Prediction

y_predicted = model.predict(x_test)
```

```python
y_predicted.shape
```

```python
y_predicted
```

```python
scaler.scale_
```

```python
scale_factor = 1/0.00682769
y_predicted = y_predicted * scale_factor
y_test = y_test *scale_factor
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test,'b',label='Original Price')
plt.plot(y_predicted, 'r', label= 'Predicted Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```
